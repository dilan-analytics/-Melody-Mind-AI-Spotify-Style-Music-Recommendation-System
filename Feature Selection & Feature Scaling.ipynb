```python
import pandas as pd

from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("spotify_clean.csv")
```

```python
df.columns
```

```python
features = [
    'danceability',
    'energy',
    'loudness',
    'speechiness',
    'acousticness',
    'instrumentalness',
    'liveness',
    'valence',
    'tempo'
]
```

```python
X = df[features]
```

```python
X.head()
```

```python
X.shape
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
scaled_df = pd.DataFrame(
    X_scaled,
    columns=features
)
```

```python
scaled_df.head()
```

```python
scaled_df.mean()
```

```python
scaled_df.std()
```

```python
scaled_df.to_csv("spotify_scaled.csv", index=False)
```
