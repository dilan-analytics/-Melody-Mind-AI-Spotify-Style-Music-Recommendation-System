```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("spotify_clean.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['popularity'], bins=20)

plt.title("Distribution of Song Popularity")

plt.xlabel("Popularity")

plt.ylabel("Number of Songs")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['danceability'], bins=20)

plt.title("Danceability Distribution")

plt.xlabel("Danceability")

plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['energy'], bins=20)

plt.title("Energy Distribution")

plt.xlabel("Energy")

plt.ylabel("Count")

plt.show()
```

```python
top_genres = df['track_genre'].value_counts().head(10)

plt.figure(figsize=(10,5))

top_genres.plot(kind='bar')

plt.title("Top 10 Genres")

plt.xlabel("Genre")

plt.ylabel("Number of Songs")

plt.show()
```

```python
top_artists = df['artists'].value_counts().head(10)

plt.figure(figsize=(12,5))

top_artists.plot(kind='bar')

plt.title("Top 10 Artists")

plt.xlabel("Artist")

plt.ylabel("Songs")

plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Matrix")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df['energy'],
    df['popularity']
)

plt.xlabel("Energy")

plt.ylabel("Popularity")

plt.title("Energy vs Popularity")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.boxplot(df['tempo'])

plt.title("Tempo Boxplot")

plt.show()
```
